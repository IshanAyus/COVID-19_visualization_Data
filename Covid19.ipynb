```python
import json
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
import plotly.io as pio
pio.renderers.default = 'browser'
```

```python
India_states = json.load(open("states_india.geojson", "r"))
India_states['features'][28]
```

```python
state_id_map = {}
for feature in India_states["features"]:
    feature["id"] = feature["properties"]["state_code"]
    state_id_map[feature["properties"]["st_nm"]] = feature["id"]
```

```python
COVID_India=pd.read_csv('covid_19_india.csv')
i=COVID_India.tail(35)
#i.to_csv('new.csv')
df=pd.read_csv('new.csv')
df.head()
```

```python
df["id"] = df["State or union territory"].apply(lambda x: state_id_map[x])
```

```python
df.head()
```

```python
df["ConfirmedScale"] = np.log10(df["Confirmed"])
```

```python
df["ConfirmedScale"].plot()
```

```python
fig = px.choropleth(
    df,
    locations="id",
    geojson=India_states,
    color="ConfirmedScale",
    hover_name="State or union territory",
    hover_data=['Confirmed','Deaths'],
    title="COVID-19 Confirmed Ratio of 12 Aug 2020",
    
)
fig.update_geos(fitbounds="locations", visible=False)
fig.show()
```

```python
df["DeathsScale"] = np.log10(df["Deaths"])
```

```python
fig = px.choropleth(
    df,
    locations="id",
    geojson=India_states,
    color="DeathsScale",
    hover_name="State or union territory",
    hover_data=['Deaths','Confirmed'],
    title="COVID-19 Deaths Ratio of 12 Aug 2020",
    
)
fig.update_geos(fitbounds="locations", visible=False)
fig.show()
```
